# Pandas

```python
import pandas as pd
```

## 시리즈 (Series)

### 1. 1번 퀘스트
- Pandas의 Series를 리스트 `[5, 10, 15, 20]`을 사용해 생성
- Series의 인덱스를 확인하는 코드 작성

```python
# 리스트를 이용한 생성
series_from_list = pd.Series([5, 10, 15, 20])

print(series_from_list)

# 인덱스 확인: Pandas Series는 기본적으로 0부터 시작하는 인덱스를 자동 생성함
print(series_from_list.index)
```

### 2. 2번 퀘스트
- 딕셔너리를 이용해 Pandas Series를 생성
- 인덱스를 활용해 'b'의 값을 출력하는 코드 작성

```python
data = {'a': 100, 'b': 200, 'c': 300}
series = pd.Series(data)

print(series)

# 인덱스 활용해 'b' 값 출력
print(series['b'])
```

### 3. 3번 퀘스트
- Pandas Series에 대해 결측값(`NaN`)을 확인
- 모든 결측값을 `0`으로 채운 후 Series의 값을 출력하는 코드 작성

```python
series = pd.Series([1, 2, None, 4, None, 6])

# 결측값 확인
print(series.isnull())

# 결측값 NaN을 0으로 대체
series_filled = series.fillna(0)

print(series_filled)
```

## 데이터프레임 (DataFrame)

### 1. 1번 퀘스트
- 데이터프레임에서 열(column)의 이름들을 출력하는 코드 작성

```python
data = {'이름': ['홍길동', '김철수', '박영희'],
        '나이': [25, 30, 28],
        '성별': ['남', '남', '여']}

df = pd.DataFrame(data)

# 열 이름 목록 반환
print(df.columns)
```

### 2. 2번 퀘스트
- 데이터프레임에서 나이 열(age)을 기준으로 오름차순으로 정렬된 새로운 데이터프레임을 생성하고 출력하는 코드 작성

```python
data = {'이름': ['홍길동', '김철수', '박영희'],
        '나이': [25, 30, 28],
        '성별': ['남', '남', '여']}

df = pd.DataFrame(data)

# 특정 열 기준 오름차순 정렬
sorted_data = df.sort_values(by='나이')

print(sorted_data)
```

### 3. 3번 퀘스트
- 각 학생의 총점을 계산하는 새로운 열을 추가한 후, 총점이 250점 이상인 학생들만 포함된 데이터프레임을 생성

```python
data = {'이름': ['홍길동', '김철수', '박영희', '이순신'],
        '국어': [85, 90, 88, 92],
        '영어': [78, 85, 89, 87],
        '수학': [92, 88, 84, 90]}

df = pd.DataFrame(data)

# 기존 열 연산
df['총점'] = df['국어'] + df['영어'] + df['수학']

# 조건에 따른 데이터 추출
high_score = df[df['총점'] >= 250]

print(high_score)
```
